```python


from ortools.sat.python import cp_model

def minimize_makespan(departments, tasks):
    model = cp_model.CpModel()

    task_vars = {}
    for task_name, task_details in tasks.items():
        for department, duration in task_details.items():
            task_vars[(task_name, department)] = model.NewIntVar(0, sum(duration for department_durations in tasks.values() for duration in department_durations.values()), f"{task_name}_{department}")

    # Constraints
    for department in departments:
        model.AddNoOverlap([task_vars[(task_name, department)] for task_name, _ in tasks.items()])

    for task_name, task_details in tasks.items():
        for department, duration in task_details.items():
            model.Add(task_vars[(task_name, department)] + duration <= task_vars[(task_name, d)] for d in departments if d != department)

    # Objective
    makespan = model.NewIntVar(0, sum(duration for department_durations in tasks.values() for duration in department_durations.values()), "makespan")
    model.AddMaxEquality(makespan, [task_vars[(task_name, department)] for task_name, task_details in tasks.items() for department in departments])
    model.Minimize(makespan)

    # Solve
    solver = cp_model.CpSolver()
    status = solver.Solve(model)

    if status == cp_model.OPTIMAL:
        optimal_schedule = {department: [] for department in departments}
        for (task_name, department), task_var in task_vars.items():
            start_time = solver.Value(task_var)
            duration = tasks[task_name][department]
            optimal_schedule[department].append((task_name, start_time, duration))

        for department in departments:
            optimal_schedule[department].sort(key=lambda x: x[1])

        return solver.ObjectiveValue(), optimal_schedule

    return None, None



departments = ["F", "J", "E", "R", "V", "B"]
tasks = {
    "Gufusofnunarkerfi": {"F": 4, "V": 24, "E": 17, "J": 30, "B": 22, "R": 27},
    "Orkuframleidslukerfi": {"F": 34, "J": 24, "E": 8, "R": 36, "V": 4, "B": 5},
    "Orkudreifingarkerfi": {"V": 32, "J": 18, "R": 26, "B": 24, "F": 10, "E": 4},
    "Mannvirki": {"B": 11, "F": 20, "R": 16, "J": 19, "E": 28, "V": 7},
}

min_makespan, optimal_schedule = minimize_makespan(departments, tasks)

print(f"Minimum makespan: {min_makespan} weeks")
print("Optimal schedule:")
for department, tasks in optimal_schedule.items():
    print(f"{department}:")
    for task_name, start_time, duration in tasks:
        print(f"  Task: {task_name}, Start Time: {start_time}, Duration: {duration}")
```
